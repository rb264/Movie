```python
import matplotlib.pyplot as plt
```

```python
accuracy_scores = [97.47, 98.77]  # Accuracy scores obtained from the notebook
```

```python
labels = ['MultinomialNB (Test Data)', 'MultinomialNB (Full Data)']
```

```python
plt.figure(figsize=(8, 5))
plt.bar(labels, accuracy_scores, color='skyblue')
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 100)

# Add the accuracy values on top of the bars
for i, v in enumerate(accuracy_scores):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')

plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(labels, accuracy_scores, marker='o', color='skyblue', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 100)

# Add data labels to the data points
for i, v in enumerate(accuracy_scores):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')

plt.grid(True)
plt.show()
```
